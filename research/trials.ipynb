#### Download Data from Gdrive

```python
import gdown
```

```python
url = 'https://drive.google.com/file/d/1z0mreUtRmR-P-magILsDR3T7M6IkGXtY/view?usp=sharing'
```

```python
type(url)
```

```python
url.split('/')
```

```python
file_id = url.split('/')[-2]
file_id
```

```python
prefix = 'https://drive.google.com/uc?export=download&id='
final_url = prefix + file_id
final_url
```

```python
gdown.download(final_url, 'chest-CT-scan-data.zip')
```
